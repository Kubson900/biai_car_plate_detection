```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    "VGG_16_batch_1": [0.83, 86.36],
    "VGG_16_batch_2": [0.83, 86.36],
    "VGG_16_batch_4": [0.93, 79.55],
    "VGG_16_batch_8": [0.83, 81.82],
    "VGG_16_batch_16": [0.82, 75.00],
    "VGG_16_batch_32": [0.84, 77.27],
    "VGG_16_batch_64": [0.95, 81.82],
    "VGG_16_batch_64_v2": [0.85, 84.09],
    "EfficientNetV2L_batch_64": [1.86, 59.09],
    "InceptionResNetV2_batch_64": [1.73, 52.27],
    "InceptionV3_batch_64": [2.28, 61.36],
}

df = pd.DataFrame.from_dict(data, orient="index", columns=["Loss", "Accuracy"])
df
```

```python
df_sorted_accuracy = df.sort_values("Accuracy")
```

```python
plt.figure(figsize=(20, 10), dpi=200)
sns.barplot(data=df_sorted_accuracy, x=df_sorted_accuracy.index, y="Accuracy")
plt.xticks(rotation=30)
plt.title("Models accuracy comparison")
plt.ylabel("Accuracy [%]")
plt.savefig("../saved_images/models_accuracy_comparison.jpg", bbox_inches="tight")
```

```python
df_sorted_loss = df.sort_values("Loss")
```

```python
plt.figure(figsize=(20, 10), dpi=200)
sns.barplot(data=df_sorted_loss, x=df_sorted_loss.index, y="Loss")
plt.xticks(rotation=30)
plt.title("Models loss comparison")
plt.ylabel("Loss [%]")
plt.savefig("../saved_images/models_loss_comparison.jpg", bbox_inches="tight")
```
